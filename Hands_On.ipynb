# 2nd approach to handle the missing values

```python
# 1. Handling the missing values with mean
```

```python
print(type(data_tips_new))
```

```python
data_tips_new
```

```python
data_tips_new.info()
```

```python
# storing all the columns name
list1 = data_tips_new.columns
list1
```

```python
print(type(data_tips_new))
```

```python
for i in list1:
    if data_tips_new[i].dtypes == 'object':
        data_tips_new[i] = data_tips_new[i].fillna(data_tips_new[i].mode()[0])
        
    else:
        data_tips_new[i] = data_tips_new[i].fillna(data_tips_new[i].mean())

```
